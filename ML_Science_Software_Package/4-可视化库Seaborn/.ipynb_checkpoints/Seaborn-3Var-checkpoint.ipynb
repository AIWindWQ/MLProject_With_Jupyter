```python
%matplotlib inline
import numpy as np
import pandas as pd
from scipy import stats, integrate
import matplotlib.pyplot as plt

import seaborn as sns
sns.set(color_codes=True)
np.random.seed(sum(map(ord, "distributions")))
```

```python
x = np.random.normal(size=100)
sns.distplot(x,kde=False)    # kde表示是否使用核密度估计
```

```python
sns.distplot(x, bins=20, kde=False)
```

### 数据分布情况  ###

```python
x = np.random.gamma(6, size=200)
sns.distplot(x, kde=False, fit=stats.gamma)  # fit表示指定其数学指标
```

根据均值和协方差生成数据

```python
mean, cov = [0, 1], [(1, .5), (.5, 1)]
data = np.random.multivariate_normal(mean, cov, 200)
df = pd.DataFrame(data, columns=["x", "y"])
df
```

###  观测两个变量之间的分布关系最好用散点图 ###

```python
sns.jointplot(x="x", y="y", data=df);   # 数据指定为DataFrame格式
```

```python
# 对于点比较多时，上面的图的使用可能会混在一起，难以分清，因此使用hex图可以改善数据点当比较多时，仍然可以直观的看到分布情况
x, y = np.random.multivariate_normal(mean, cov, 1000).T
with sns.axes_style("white"):
    sns.jointplot(x=x, y=y, kind="hex", color="k")  # hex图
```
